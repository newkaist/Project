```python
import os
import glob
```

```python
s_path = './Item-Pose_f/*'
```

```python
glob.glob(s_path)[0].split('\\')[1].split('.')[0]
```

```python
file_list = glob.glob(s_path)
```

```python
file_list[:10]
```

```python
# 파일 명 리스트로 만들기
file_name_list = [file_list[i].split('\\')[1].split('.')[0] for i in range(len(file_list))]
```

```python
len(file_name_list), file_name_list[:10]
```

```python
# category_name 저장
import json

file_cat=[]

for filename in file_name_list:
    file_path = f'./Item-Pose_f/{filename}.json'
    with open(file_path, 'r') as file:
        data = json.load(file)
        file_cat.append([filename, data["category_name"]])
```

```python
len(file_cat), file_cat[:10]
```

```python
import pandas

df = pandas.DataFrame(file_cat, columns = ['filename', 'cat'])
```

```python
df.head()
```

```python
df.cat.unique()
```

```python
# 이미지 파일 라벨 기준 분류 작업
import shutil

for (filename, cat) in zip(df.filename, df.cat):
        try:
            if (cat == 'cap_and_hat') or (cat == 'shoes'):
                file_path = f'./Item-Image/{filename}.jpg'
                os.remove(file_path)
            else:
                chpath = f'./Item-Image/{cat}/'
                if not os.path.exists(chpath):
                    os.makedirs(chpath)
                shutil.move('./Item-Image/'+f'{filename}.jpg', chpath+f'{filename}.jpg')
        except:
            pass
```
